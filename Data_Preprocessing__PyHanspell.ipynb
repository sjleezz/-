## PyHanspell - Naver 맞춤법 수정 API를 이용한 패키지

```python
import os
from hanspell import spell_checker
from hanspell.constants import CheckResult
```

### Insert Sentences

```python
os.getcwd()
```

```python
DIR = os.getcwd()
names = ['newskbs', 'sbs8news', 'MBCNEWS11']
filepaths = [os.path.join(DIR, name) for name in names]
paragraphs = []
for filepath in filepaths :
    files = os.listdir(filepath)
    for file in files :
        s = ''
        if file[-3:] == 'txt':
            fpath = os.path.join(filepath, file)
            with open(fpath, 'r', encoding='utf8') as f :
                lines = f.readlines()
                for line in lines :
                    s += line      
            paragraphs.append(s)
            #print(s, end='\n\n')    
print(paragraphs)            
```

### parse lines - KSS

```python
import kss
```

```python
stcs = []
for p in paragraphs :
    stc = []
    for s in kss.split_sentences(p):
        stc.append(s)
        print(s, end='\n\n')
    stcs.append(stc)    
```

### Corrected Sentences

```python
for stc in stcs :
    for s in stc :
        spelled_sent = spell_checker.check(s)
        hanspell_sent = spelled_sent.checked
        print(hanspell_sent, end='\n\n')
```

### Analysis

```python
for stc in stcs :
    for s in stc :
        spelled_sent = spell_checker.check(s)
        #print(spelled_sent.as_dict())
        for k, v in spelled_sent.words.items() :
            if v == 0 :
                continue
            print(k, v)
```
