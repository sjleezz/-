## PyHanspell - Naver 맞춤법 수정 API를 이용한 패키지

```python
import os
from hanspell import spell_checker
from hanspell.constants import CheckResult
```

### Insert Sentences

```python
os.getcwd()
```

```python
DIR = os.getcwd()
names = ['newskbs', 'sbs8news', 'MBCNEWS11']
filepaths = [os.path.join(DIR, name) for name in names]
for filepath in filepaths :
    files = os.listdir(filepath)
    for file in files :
        if file[-3:] == 'txt':
            sentence = ''
            fpath = os.path.join(filepath, file)
            with open(fpath, 'r', encoding='utf8') as f :
                lines = f.readlines()
                for line in lines :
                    sentence += line            
            print(sentence)            
```

### parse lines

```python
cnt = 0
n = 10
start = 0
sentences = []
for i, sent in enumerate(sentence):
    if sentence[i] == '니' :
        if sentence[i+1] == '다' or sentence[i+1] == '까' :
            sent = sentence[start:i+2]
            sentences.append(sent)
            print(sent) 
            print()
            cnt+=1 
            start = i+2
            #if cnt >= n :
            #    break
               
```

### Corrected Sentences

```python
for s in sentences :
    spelled_sent = spell_checker.check(s)
    hanspell_sent = spelled_sent.checked
    print(hanspell_sent)
    print()
```

### Analysis

```python
for s in sentences :
    spelled_sent = spell_checker.check(s)
    #print(spelled_sent.as_dict())
    for k, v in spelled_sent.words.items() :
        if v == 0 :
            continue
        print(k, v)
```
