## PyKoSpacing

```python
import os
from pykospacing import Spacing
```

### Insert Sentences

```python
DIR = os.getcwd()
names = ['newskbs', 'sbs8news', 'MBCNEWS11']
filepaths = [os.path.join(DIR, name) for name in names]
paragraphs = []
for filepath in filepaths :
    files = os.listdir(filepath)
    for file in files :
        s = ''
        if file[-3:] == 'txt':
            fpath = os.path.join(filepath, file)
            with open(fpath, 'r', encoding='utf8') as f :
                lines = f.readlines()
                for line in lines :
                    s += line      
            paragraphs.append(s)
            #print(s, end='\n\n')    
print(paragraphs)            
```

### Original sentences

```python
new_sents=[]
for p in paragraphs :
    no_white = p.replace(' ', '')
    new_sents.append(no_white) # 띄어쓰기가 없는 문장 임의로 만들기
    print(no_white, end='\n\n')
```

### Analyzed Sentences

```python
for s in new_sents :
    spacing = Spacing()
    kospacing_sent = spacing(s) 

    print(s, end='\n')
    print(kospacing_sent, end='\n\n')
```
